```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer los datos del archivo .txt
df = pd.read_csv('pitch3.txt', delimiter=',')

# Graficar las columnas (ajusta los nombres si es necesario)
df.plot(x='Time_s', y='F0_Hz', kind='line')
plt.xlabel('Time[S]')
plt.ylabel('Frec[Hz]')
plt.title('Pitch')
plt.show()

```

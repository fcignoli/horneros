```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer los datos del archivo .txt
df = pd.read_csv('pitch_mio.txt', delimiter=',')
time_felipe = df['Time_s']
frec_felipe = df['F0_Hz']

df2 = pd.read_csv('pitch_tomas.txt', delimiter=',')
time_tomas = df2['Time_s']
frec_tomas = df2['F0_Hz']


# Graficar las columnas (ajusta los nombres si es necesario)
plt.figure(figsize=(25, 10))
#df.plot(x='Time_s', y='F0_Hz', kind='line')
plt.plot(time_felipe, frec_felipe, label='Felipe')
plt.plot(time_tomas, frec_tomas, label='Toms')
plt.xlabel('Time[S]')
plt.ylabel('Frec[Hz]')
plt.legend()
plt.title('Pitch')
plt.show()

```
